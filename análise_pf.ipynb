Para realização do tratamento dos dados o primeiro passo é importar as bibliotecas que auxiliam no desenvolvimento do projeto. Para este caso, serão utilizados; pandas, matplotilib, seaborn.

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

Em seguida, é preciso ler a base de dados disponibilizada no site de micro dados apresentado pelo governo federal, por meio do ministério da educação, pelo portal da SAEB(sistema de avaliação de ensino básico).

```python
df_região_EM=pd.read_csv('TS_REGIAO.csv', sep=';')
df_região_EM
```

Após ler o conjunto de dados, inicia-se, de fato o tratamento.

Verficando se há valores nulos e se estes valores afetarão a análise.

```python
df_região_EM.isnull().sum()
```

Excluindo colunas que não serão necessárias.

```python
df2=df_região_EM.drop(columns=['nivel_0_MTEMT', 
                               'nivel_1_MTEMT','nivel_2_MTEMT',
                               'nivel_3_MTEMT','nivel_4_MTEMT', 
                                'nivel_5_MTEMT','nivel_6_MTEMT',
                                'nivel_7_MTEMT', 'nivel_8_MTEMT', 
                                'nivel_9_MTEMT','nivel_10_MTEMT', 
                                'nivel_0_MTEMI', 'nivel_7_MTEM',
                                'nivel_8_MTEM','nivel_9_MTEM',
                                'nivel_10_MTEM', 'nivel_8_MTEMI',
                                'nivel_9_MTEMI','nivel_10_MTEMI',
                                'nivel_0_MTEM','nivel_1_MTEM',
                                'nivel_2_MTEM','nivel_3_MTEM',
                                'nivel_4_MTEM','nivel_5_MTEM',
                                'nivel_6_MTEM','nivel_1_MTEMI',
                                'nivel_2_MTEMI','nivel_3_MTEMI',
                                'nivel_4_MTEMI','nivel_5_MTEMI',
                                'nivel_6_MTEMI','nivel_7_MTEMI'])
df2
```

Renomeando uma coluna que possa ser a intersecção que unirá os datasets no final.

```python
df2.rename(columns={'NO_REGIAO': 'REGIAO'}, inplace=True)
```

Após verificar e apagar as colunas indesejáveis, serão excluido as entradas com valores nulos dos dados restantes.

```python
df2.dropna(inplace=True)
```

Observando de que tipo são as colunas do datasets para aplicar algumas funções do pandas.

```python
df2.info()
```

Nese caso, como as colunas estão descrtias como OBJECT é como se o python não entendesse com precisão qual o tipo de informação, desse modo, é necessário que se altera o tipo da descrição de algumas colunas, para isto será aplicado alguns ajustes para que a função pd.to_numeric realize a troca do tipo de informação.

```python
df2['MEDIA_2_MT'] = df2['MEDIA_2_MT'].str.replace(r'[^\d.]+', '', regex=True)
```

```python
df2['MEDIA_2_MT'] = pd.to_numeric(df2['MEDIA_2_MT'], errors='coerce')
```

Ao trocar o tipo da coluna os valores perdem a precisão com relação a casas decimais. Por isto, é necessário uma normalização das notas.

```python
df2['MEDIA_2_MT'] = df2['MEDIA_2_MT'].apply(lambda x: int(x /100))
```

***ANÁLISE***

Feito as normalizações, ajustes e limpeza dos dados, incia-se as análise dos dados tratados.
Para estes momento, a ideia é observar qual a média obtida pelos alunos do ensino médio de cada região na área de matemática e suas tecnologias.

```python
df_mediaregiao = df2.groupby('REGIAO').agg({'MEDIA_2_MT':'mean'}).round()
df_mediaregiao
```

**Visualização dos dados**

```python
plt.figure(figsize=(6, 4))


contagem = df_mediaregiao .groupby('REGIAO')['MEDIA_2_MT'].mean()
contagem.plot.bar()


plt.title('Média em matemática por região')
plt.xlabel('Região')
plt.ylabel('Notas')
plt.show()
```

***Tratando o segundo dataset para fazer um merge visando a criação de um novo dataset, que por sua vez será fonte para a criação de um dashboard***

Seguindo os passo anteriormente citados, primeiramente é necessário ler o dataset. Este, disponilizado pelo MEC(ministério da educação e da cultura).

```python
dfzin=pd.read_csv('PDA_Dados_Cursos_Graduacao_Brasil.csv')
dfzin
```

Excluindo colunas despensáveis.

```python
dff=dfzin.drop(columns=['CODIGO_IES','SITUACAO_CURSO',
                        'ORGANIZACAO_ACADEMICA','CODIGO_CURSO',
                        'CARGA_HORARIA','CODIGO_AREA_OCDE_CINE','CODIGO_MUNICIPIO'])
dff
```

Analisando as entradas com valores nulos e se elas inteferem na ánalise.

```python
dff.isnull().sum()
```

```python
dff.info()
```

Agrupando os dados para por meio da função groupby e agg para que seja possível relizar operações com os valores de entrada.
O objetivo aqui é de visulizar o número de vagas dispostas por cada região. Vale ressaltar que, essas vagas são distribuidas em todos os cursos ofertados pelas universidades. 

```python
df_quant_vagasgeral = dff.groupby('REGIAO').agg({'QT_VAGAS_AUTORIZADAS':'size'})
df_quant_vagasgeral
```

Aqui, agrupa-se a quantidade de vagas para cada área tipo de curso, pois o que se quer aqui são as vagas nos cursos de física.

```python
df_quant_vagas_ac = dff.groupby('AREA_OCDE').agg({'QT_VAGAS_AUTORIZADAS':'size'})
df_quant_vagas_ac
```

Isolando apenas as vagas no curso de física.

```python
df_filtrado = dff[dff['AREA_OCDE'].str.contains('Física',  na=False)]
df_filtrado
```

Agrupando e calculandoo numero de vagas preenchidas para o curso de Física para cada região.

```python
df_quant_vagas_fsc = df_filtrado.groupby(['AREA_OCDE', 'REGIAO']).agg({'QT_VAGAS_AUTORIZADAS':'size'})
df_quant_vagas_fsc
```

**Visualização**


Projeção gráfica do número de vagas preenchidas por cada região.


```python
plt.figure(figsize=(6, 4))


contagem = df_quant_vagas_fsc.groupby('REGIAO')['QT_VAGAS_AUTORIZADAS'].mean()
contagem.plot.bar()


plt.title('Número de vagas em Física por região')
plt.xlabel('Região')
plt.ylabel('Quantidade de vagas')
plt.show()
```

**Após os tratamentos e análises, hora de fazer um novo dataset com os dados dos datassets tratados, por meio da função MERGE**

```python
dffinal= pd.merge(dff,df2, on=['REGIAO'], how='outer')
dffinal
```
